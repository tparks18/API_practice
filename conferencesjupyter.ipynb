```python
import json
import requests

url = 'http://ct-api-challenge.herokuapp.com/'
data = requests.get(url).json()

# print(data.keys())
# print(len(data))
# print(len(data['partners']))

#country = data['partners'][i]['country']

# {'availableDates': ['2017-06-05', '2017-06-06', '2017-06-12', '2017-06-13', '2017-06-16', '2017-06-17', '2017-06-25'],
#     'country': 'France', 'email': 'bmerrithew@codingtemple.com', 'firstName': 'Benton', 'lastName': 'Merrithew'}

availableDates = {}
# creates country dictionary with emails associated with dates
for i in range(len(data['partners'])):
    # country
    country = data['partners'][i]['country'] #gave country a variable
    if country not in availableDates:
        # This makes another dictionary to place every country in another dictionary
        availableDates[country] = {}

    # Dates each partner is available for set to a variable
    partnerDates = data['partners'][i]['availableDates']
    for date in partnerDates:
        if date not in availableDates[country]:
            availableDates[country][date] = [data['partners'][i]['email']]
        else:
            availableDates[country][date].append(data['partners'][i]['email'])
# 18-31 loops through the original data set once

# print(availableDates)
# for country in availableDates:
#     print(country)

#print(availableDates['United States'])

# iterate through countries and find largest date array
mostAttendanceByStartDate = 0  # highest number of people for any date
bestDates = []  # this for loop calculates the best dates for each country based on how many people can attend
for country in availableDates:
    for dateList in availableDates[country]:
        #print(len(availableDates['United States'][dateList]))
        if len(availableDates[country][dateList]) >= mostAttendanceByStartDate:
            mostAttendanceByStartDate = len(availableDates[country][dateList])
            bestDates.append(str(dateList))
    print(country)
    # Using the for loop above, the dates that get printed are are in order of least to most attendance
    print(bestDates[-2:])
    
# the -2 allows us to see what the last two dates are in the list, giving us the maximum amount of attendees
#print(availableDates[country][date])
```
